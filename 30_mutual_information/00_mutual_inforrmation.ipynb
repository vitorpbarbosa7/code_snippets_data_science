# Entropy

```python
import math
from typing import Tuple, Union
```

```python
math.log(0.2,2)
```

```python
from functools import reduce
```

```python
args = (0.2,0.8)
reduce(lambda x, y: -x*math.log(x,2) + -y*math.log(y,2), args)
```

```python
def entropy(*args) -> float:
    return reduce(lambda x, y: -x*math.log(x,2) + -y*math.log(y,2), *args)
```

```python
entropy(0.2,0.8)
```

## Balanced case

```python
entropy((0.25,0.25,0.25,0.25))
```

```python
probs = tuple([0.2]*5)
entropy(probs)
```

## Unbalanced case

```python
probs = (0.2,0.8)
entropy(probs)
```

```python
probs = np.arange(0.01,0.99,0.01)
tupleprobs = tuple(zip(probs, 1-probs))
```

```python
fg(8,4)
plt.scatter(x = probs, y = [entropy(prob) for prob in tupleprobs])
```

## Machine Learning Mastery

https://machinelearningmastery.com/information-gain-and-mutual-information/

```python
# calculate the entropy for a dataset
from math import log2
# proportion of examples in each class
class0 = 10/100
class1 = 90/100
# calculate entropy
entropy = -(class0 * log2(class0) + class1 * log2(class1))
# print the result
print('entropy: %.3f bits' % entropy)
```

# Mutual Information at last

- Em outros termos, qual a reduo na entropia de uma varivel aleatria, dado o conhecimento sobre a distribuio de probabilidade de outra varivel
